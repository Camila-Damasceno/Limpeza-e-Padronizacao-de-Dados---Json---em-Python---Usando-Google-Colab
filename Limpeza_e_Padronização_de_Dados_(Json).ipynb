```python
#carregamento de dados (json)

df_customers = pd.read_json("/content/customers_data.json")
df_products = pd.read_json("/content/products_data.json")
df_sales = pd.read_json("/content/sales_data.json")
```

#**Limpeza de Dados**

```python
#limpeza de dados - customers

#para dados de idade ausentes
media_idade = df_customers['age'].mean()

#substituir o campo
#chamar o campo e indicar que ele vai substituir ele mesmo com condições
df_customers['age'] = df_customers['age'].fillna(media_idade)


#para data inválida
df_customers["signup_date"] = pd.to_datetime(df_customers["signup_date"], errors="coerce")

#padronizar gênero
#replace - o que era e o que vai ser
#str.upper - para deixar tudo maiúsculo
df_customers['gender'] = (df_customers["gender"].replace("",pd.NA).str.upper())

display (df_customers)
```

```python
#limpeza dados - product
#trocar base_cost vazio por média do base_cost
media_base_cost = df_products['base_cost'].mean()

df_products['base_cost'] = df_products['base_cost'].fillna(media_base_cost)

display(df_products)
```

```python
#limpeza dados - sales
#converter tipo de dados (dtype)

df_sales['sales'] = pd.to_numeric(df_sales['sales'], errors = 'coerce')

#localizar valor de vendas <= 0
#1 - chama a base de dados com o atributo .loc - localizar
#2 - o campo desejado
#3 - informa a condição e o campo
#4 - informa o que vai substituir

df_sales.loc[df_sales['sales']<=0, 'sales'] = pd.NA

#converter datas

df_sales['date'] = pd.to_datetime(df_sales['date'], errors = 'coerce')
print("quantidade de dados truncados")
print(df_sales['date'].isna().sum())

display(df_sales)

#resolver dados errados em região

df_sales['region'] = df_sales['region'].str.strip().str.title()

#resolver problemas no customer_type

df_sales['customer_type'] = df_sales['customer_type'].replace('',pd.NA)
```

```python
#ATENÇÃO - COMANDO DE DELETE
df_sales_dados_limpos = df_sales.dropna(subset=[]) #CAMPOS A SEREM DELETADOS
#apaga do data set
```

**Tabela Dinâmica** / **Tabela de Correlações**

```python
#criar campo de correlação entre tabelas

mapeamento_clientes = {
    'Novo': "C001",
    'Recorrente': "C002",
    'Corporativo': "C003"
}

#adicionar novo campo à tabela

df_sales['customer_id'] = df_sales['customer_type'].map(mapeamento_clientes)

df_dados_consolidados = df_sales.merge(df_customers, on="customer_id")

display(df_dados_consolidados)
```

```python
#renomear campos

df_dados_consolidados = df_dados_consolidados.rename(columns={'region_x': 'region_sale'})
df_dados_consolidados = df_dados_consolidados.rename(columns={'region_y': 'region_customer'})

display(df_dados_consolidados)
```

```python
#pivot_table

#quantidade de vendas por regiao

pd.pivot_table(
    # 1 - qual base de dados
    df_dados_consolidados,
    # 2 - Quais valores
    values="quantity",
    # 3 - quais os index (legendas)
    index = 'region_sale',
    # função de agregação
    aggfunc="sum"
)
```

```python
#Ticket médio por região
# correlacione o valor de venda por regiao
pd.pivot_table(
    df_dados_consolidados,
    values=['sales', 'cost'],
    index = "region_sale",
    aggfunc="mean"
)
```
